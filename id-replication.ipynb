```python
class Base_FormulasAndFunctions:
    """
    This class provides formulas and functions commonly used for variables.

    Attributes:
    - df (pd.DataFrame): The last DataFrame to be processed.
    - result (pd.DataFrame): The last result.
    - _preprocessed_df (pd.DataFrame): The last preprocessed DataFrame.
    """
    def __init__(self):
        self.df: pd.DataFrame = None  # The last df to be processed
        self.result: pd.DataFrame = None  # The last result
        self._preprocessed_df: pd.DataFrame = None  # The last preprocessed df

    ''' 
    Below are the formulas to calculate commonly used variables.
    ''' 
    @staticmethod
    def _avg_price(price: pd.Series) -> pd.Series:
        return price.mean()

    @staticmethod
    def _dollar_value(price, size):
        return price * size

    @staticmethod
    def _dollar_value_sum(group):
        return (group["price"] * group["size"]).sum()

    @staticmethod
    def _volume_sum(group):
        return group["size"].sum()

    @staticmethod
    def _bid_ask_mid_price(nbb: str, nbo: str) -> pd.Series:
        return (nbb + nbo) / 2

    @staticmethod
    def _calculate_d_k(buy_sell_col: pd.Series) -> pd.Series:
        return np.where(buy_sell_col.isin(['Buy', 'buy']), 1, -1)

    @staticmethod
    def _calculate_pes(price_col: pd.Series, midprice_col: pd.Series, buy_sell_id: pd.Series) -> pd.Series:
        d_k_col = Base_FormulasAndFunctions._calculate_d_k(buy_sell_id)
        return (2 * d_k_col * (price_col - midprice_col)) / midprice_col

    @staticmethod
    def _calculate_des(price_col: pd.Series, midprice_col: pd.Series, buy_sell_id: pd.Series) -> pd.Series:
        d_k_col = Base_FormulasAndFunctions._calculate_d_k(buy_sell_id)
        return 2 * d_k_col * (price_col - midprice_col)

    @staticmethod
    def _calculate_dollar_weight_multiplier(price_col: pd.Series, size_col: pd.Series) -> pd.Series:
        dv = Base_FormulasAndFunctions._dollar_value(price_col, size_col)
        return dv / dv.sum()

    @staticmethod
    def _calculate_volume_weight_multiplier(size_col: pd.Series) -> pd.Series:
        return size_col / size_col.sum()

    @staticmethod
    def _calculate_dollar_value_weighted_price(group, price_col: str, size_col: str):
        return group[price_col] * (
                Base_FormulasAndFunctions._calculate_dollar_weight_multiplier(group[price_col], group[size_col]))

    @staticmethod
    def _calculate_volume_weighted_price(group, price_col: str, size_col: str) -> pd.Series:
        return group[price_col] * (Base_FormulasAndFunctions._calculate_volume_weight_multiplier(group[size_col]))

    @staticmethod
    def _calculate_vw_avg_price(group, price_col: str, size_col: str) -> pd.Series:
        return (group[price_col] * (
                Base_FormulasAndFunctions._calculate_volume_weight_multiplier(group[size_col]))).mean()
    
    def _get_dropclones_intraday(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        Preprocesses the given DataFrame by filtering for regular market hours and dropping clone rows.

        Parameters:
        - df (pd.DataFrame): The DataFrame containing TAQ trades and NBBO quotes.

        Returns:
        - df (pd.DataFrame): The preprocessed DataFrame with regular market hours and clone rows dropped.
        """

        # get intraday based on available market open and close trade conditions
        if df["tr_scond"].str.contains("O").any() and df["tr_scond"].str.contains("6").any():
            # Filter for rows with "O" and "6" in the "tr_scond" column
            dt_open = df[df["tr_scond"].str.contains("O")]['datetime']
            dt_close = df[df["tr_scond"].str.contains("6")]['datetime']

            # Convert datetime values to string format
            dt_open = pd.to_datetime(dt_open).dt.strftime('%H:%M:%S.%f')
            dt_close = pd.to_datetime(dt_close).dt.strftime('%H:%M:%S.%f')

            # Extract start and end times from the first row
            start_time = pd.to_datetime(dt_open.iloc[0]).time()
            end_time = pd.to_datetime(dt_close.iloc[0]).time()

            # Filter DataFrame for rows within the specified time range
            df = df.set_index("datetime").between_time(start_time, end_time, inclusive="both").reset_index(drop=False)
        else:
            # If "O" and "6" are not present, filter for regular market hours (9:30 to 16:00)
            df = df.set_index("datetime").between_time("9:30", "16:00", inclusive="left").reset_index(drop=False)

        # Drop clone rows
        six_rows = df[df['tr_scond'].str.contains("6")]
        M_rows = df[df['tr_scond'].str.contains("M")]

        o_rows = df[df['tr_scond'].str.contains("O")]
        q_rows = df[df['tr_scond'].str.contains("Q")]

        M_rows = M_rows[M_rows['size'].isin(six_rows['size'].values)]
        M_rows = M_rows[M_rows['price'].isin(six_rows['price'].values)]

        q_rows = q_rows[q_rows['size'].isin(o_rows['size'].values)]
        q_rows = q_rows[q_rows['price'].isin(o_rows['price'].values)]

        df = df[~(df.index.isin(M_rows.index))]
        df = df[~(df.index.isin(q_rows.index))]

        return df
    
    def _general_preprocess(self, df):
        """
        This function performs general preprocessing on the input data.

        Parameters:
        input_data (type): Description of the input_data parameter. Replace 'type' with the actual data type.

        Returns:
        output_data (type): Description of the output_data that the function returns. Replace 'type' with the actual data type.

        Raises:
        ExceptionType: Description of the circumstances under which the function raises this exception. Replace 'ExceptionType' with the actual type of exception.

        """
        assert set(['ticker', 'datetime']).issubset(df.columns), "The dataframe must contain 'ticker' and 'datetime' columns."
        # Rememeber the last processed df
        self.df = df
        df = df.copy()

        # Setting datetime index

        # Filtering for regular market hours
        df = df.groupby(['ticker', 'date']).apply(self._get_dropclones_intraday).reset_index(drop=True)
        return df


```

```python
""" This module implements summaries from WRDS Intraday Indicator Data.
"""

# Formula 1 Lee and Ready 1991 (Modified for tick test in formula 2)
def lee_ready_1991_trade_clf(df: pd.DataFrame, price: str, ba_mid_price: str, groupby_identifier: pd.Grouper, tick_test: str=None) -> pd.Series:
    lagged_ba_price = groupby_identifier[ba_mid_price].shift(1)

    if tick_test:
        buy_mask = (df[price] > lagged_ba_price) | (df[tick_test].isin(["uptick", "zero_uptick"]) & df[price] == lagged_ba_price)
        sell_mask = (df[price] < lagged_ba_price) | (df[tick_test].isin(["downtick", "zero_downtick"]) & df[price] == lagged_ba_price)
    else:
        buy_mask = df[price] > lagged_ba_price
        sell_mask = df[price] < lagged_ba_price
        tick_test_mask  = df[price] == lagged_ba_price

    lr_trade_clf = pd.Series(
        np.where(buy_mask, "Buy", np.where(sell_mask, "Sell", np.nan)),
        index=df.index
    )
    if not tick_test:
        lr_trade_clf.loc[tick_test_mask] = "tick_test"
    
    return lr_trade_clf

# Formula 2 Lee and Ready 1991 Tick Test
def lee_ready_1991_tick_test(df: pd.DataFrame, price: str, groupby_identifier: pd.Grouper) -> pd.Series:
    lagged_trade_price = groupby_identifier[price].shift(1)

    uptick_mask = df[price] > lagged_trade_price
    downtick_mask = df[price] < lagged_trade_price

    df["tick_test"] = pd.Series(
        np.where(uptick_mask, "uptick", np.where(downtick_mask, "downtick", np.nan)),
        index=df.index
    )
    lagged_tick_test = groupby_identifier["tick_test"].shift(1)
    zero_uptick_mask = (df[price] == lagged_trade_price) & (lagged_tick_test == "uptick")
    zero_downtick_mask = (df[price] == lagged_trade_price) & (lagged_tick_test == "downtick")

    lr_tick_test = pd.Series(
        np.where(zero_uptick_mask, "zero_uptick", np.where(zero_downtick_mask, "zero_downtick", df["tick_test"])),
        index=df.index
    )
    return lr_tick_test

#preliminary calculations for the rest of the variables
def calc_columns(dataframe1, buy_sell_clf: str):
    dataframe = dataframe1.copy()
    
    
    #first vector calculate rows that aren't side specific.
    dataframe["ba_midprice"] = (dataframe['nbb']+dataframe['nbo'])/2
    # dataframe["ba_midprice"] = dataframe["ba_midprice"].shift(-1) #I tried this -1 and 1, bad results.
    
    
    dataframe["dollar_value"] = dataframe['size']*dataframe['price']
    dataframe["w_k_dollar_all"] = dataframe['dollar_value']/(dataframe['dollar_value'].sum())
    dataframe["w_k_shares_all"] = dataframe['size']/(dataframe['size'].sum())
    # dataframe['w_k_dollar*p_k_all'] = dataframe["w_k_dollar_all"]*dataframe["price"]

    #side specifc
    dataframe["w_k_dollar"] = 0
    dataframe['w_k_dollar*p_k'] = 0
    dataframe["w_k_shares"] = 0

    
    df_bs = dataframe.groupby(buy_sell_clf)
    df_buy = df_bs.get_group('Buy')
    df_sell = df_bs.get_group('Sell')    
    # d_k = len(df_buy)-len(df_sell)
    
    ###buy specific
    
    df_buy["w_k_dollar"] = df_buy['dollar_value']/(df_buy['dollar_value'].sum()) #when u use w_k_dollar_all for value weighted avg price - something side specific it is bad
    df_buy["w_k_shares"] = df_buy['size']/(df_buy['size'].sum())
    df_buy['w_k_dollar*p_k'] = df_buy["w_k_dollar"]*df_buy["price"]

    ###sell specific
    df_sell["w_k_dollar"] = df_sell['dollar_value']/(df_sell['dollar_value'].sum())
    df_sell["w_k_shares"] = df_sell['size']/(df_sell['size'].sum())
    df_sell['w_k_dollar*p_k'] = df_sell["w_k_dollar"]*df_sell["price"]
    
    dataframe.loc[dataframe[buy_sell_clf] == ('Buy' or 'buy'), :] = df_buy.values
    dataframe.loc[dataframe[buy_sell_clf] == ('Sell' or 'sell'), :] = df_sell.values

    #using D_k
    dataframe['d_k'] = np.where(dataframe[buy_sell_clf] == ('Buy' or 'buy'), 1, -1)
    dataframe['p-m'] = dataframe["price"]-dataframe["ba_midprice"]
    dataframe['PES'] = (2*dataframe['d_k']*dataframe['p-m'])/dataframe["ba_midprice"]
    #Simple Averaged Dollar Effective Spread Formula 3
    dataframe['DES'] = 2*dataframe['d_k']*dataframe['p-m']

    #requiring end_time = current_time + pd.Timedelta(minutes=5)

    # calc_Dollar_Pct_RS_PI_Calculator(dataframe)


    return dataframe

# Formula 2 Lee and Ready 1991 Tick Test
def WRDS_trade_clf(df: pd.DataFrame, price: str, LR_clf: str, tick_test_clf: str) -> pd.Series:
    # WRDS_clf = []
    in_nbbo = (df[price] >= df['nbb']) and (df[price] <= df['nbo'])
    df["WRDS_clf"] = np.where(in_nbbo, df[LR_clf], df[tick_test_clf])

    return df["WRDS_clf"]

```

```python
class SpecialOrderSummary(Base_FormulasAndFunctions):
    
    """
    This class implements Special Order Summaries.
    """
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self._preprocessed_df_sweep: pd.DataFrame = None  # The first preprocessed df
        self._preprocessed_df_oddlot: pd.DataFrame = None  # The second preprocessed df


    def _preprocess(self, df: pd.DataFrame):
        """
        Preprocess the data by setting a datetime index and filtering for regular market hours.

        Parameters:
        - df (pd.DataFrame): A pandas DataFrame containing TAQ trades and NBBO quotes.
        """
        # Apply general preprocessing
        self._preprocessed_df = self._general_preprocess(df)
        
        # Filter for odd lot and sweep trades
        df_oddlot = self._preprocessed_df[self._preprocessed_df['tr_scond'].str.contains("I")]
        df_sweep = self._preprocessed_df[self._preprocessed_df['tr_scond'].str.contains("F")]

        self._preprocessed_df_oddlot = df_oddlot.groupby(['ticker', 'date'])
        self._preprocessed_df_sweep = df_sweep.groupby(['ticker', 'date'])
        
        

    def n_iso_trades(self, df: pd.DataFrame) -> pd.Series:
        """ The Number of Intermarket Sweep Order Trades (During Market Hours).

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The number of ISO trade for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = self._preprocessed_df_sweep.size()
        return self.result


    def iso_dollar(self, df: pd.DataFrame) -> pd.Series:
        """ Sum of Intermarket Sweep Order Trade Dollar Value (During Market Hours)

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The sum of dollar value for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = self._preprocessed_df_sweep.apply(self._dollar_value_sum)
        return self.result

    def iso_vol(self, df: pd.DataFrame) -> pd.Series:
        """ Sum of Intermarket Sweep Order Trade Dollar Value (During Market Hours)

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The sum of dollar value for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = self._preprocessed_df_sweep.apply(self._volume_sum)
        return self.result

    def n_oddlot_trades(self, df: pd.DataFrame) -> pd.Series:
        """ Number of Odd Lot Trades (During Market Hours)

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The number of odd lot trades for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = self._preprocessed_df_oddlot.size()
        return self.result

    def oddlot_dollar(self, df: pd.DataFrame) -> pd.Series:
        """ Sum of Odd Lot Trade Value (Dollar Volume during Market Hours)

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The sum of dollar value for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = self._preprocessed_df_oddlot.apply(self._dollar_value_sum)
        return self.result

    def oddlot_vol(self, df: pd.DataFrame) -> pd.Series:
        """ Sum of Odd Lot Trade Volume (During Market Hours)

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The sum of odd lot trade volume for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result =  self._preprocessed_df_oddlot.apply(self._volume_sum)
        return self.result

    def n_outside_nbbo(self, df: pd.DataFrame) -> pd.Series:
        """ Number of trades out of the previous nanosecond National Best Bid-Ask range

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The number of trades outside the NBBO for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = df.apply(lambda df: df.apply(lambda row: row['price'] > row['nbo'] or row['price'] < row['nbb'], axis=1).sum(), axis=1)
        return self.result

    def ret_mkt_m(self, df: pd.DataFrame) -> pd.Series:
        """ Log Open to Close Return (During Market Hours)

        When official market opening trade or official market closing trade are missing, set to missing.
        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: The log return for each group.
        :rtype: pd.Series
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)
        self.result = df.apply(lambda df: np.log(df['price'].iloc[0] / df['price'].iloc[-1]), axis=1)
        return self.result

    def generate_sos_summaries(self, df: pd.DataFrame) -> pd.DataFrame:
        """ Generate a dataframe of all the Special Order Summary statistics, sorted by ticker and date.

        :param df: If provided, the method will be applied to the dataframe instead of the original dataframe.
        :type df: pd.DataFrame
        :return: A dataframe of all the Special Order Summary statistics.
        :rtype: pd.DataFrame
        """
        if not df.equals(self.df) or self._preprocessed_df_sweep is None:
            self._preprocess(df)

        # Create a dictionary of all the Special Order Summary statistics
        sos_dict = {'n_iso_trades': self.n_iso_trades(df),
                    'iso_dollar': self.iso_dollar(df),
                    'iso_vol': self.iso_vol(df),
                    'n_oddlot_trades': self.n_oddlot_trades(df),
                    'oddlot_dollar': self.oddlot_dollar(df),
                    'oddlot_vol': self.oddlot_vol(df),
                    'n_outside_nbbo': self.n_outside_nbbo(df),
                    'ret_mkt_m': self.ret_mkt_m(df)}

        # Convert the dictionary to a dataframe
        sos_df = pd.DataFrame(sos_dict)

        return sos_df
    
```
